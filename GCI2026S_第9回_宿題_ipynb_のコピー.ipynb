```python
# common
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# googleドライブから読み込む(自分の環境に合わせて要修正)
path_winequality_data = "/content/drive/MyDrive/Colab_Notebooks/winequality-red.csv"
```

```python
# Omnicampusに提出するのはこのセルのみ
def homework(path_winequality_data, n):
  #!!WRITE ME!!
    df = pd.read_csv(path_winequality_data, sep = ';')
    path_winequality_data = df.values
    scaler = StandardScaler()
    Z_score = scaler.fit_transform(df)

    model_kmeans = KMeans(n_clusters = n, init = 'k-means++',random_state = 0)# WRITE ME
    model_kmeans.fit_predict(Z_score)

    cluster_counts  = np.bincount(model_kmeans.labels_)
    return cluster_counts
```

```python
# 確認のためのコード
n=3  #任意の正の整数に設定してください
# 戻り値は、array([130, 235, 291,  44, 268,  28, 128, 157, 151, 167]) となる

homework(path_winequality_data, n)
```
